````python
import pickle
````

````python
data = pickle.load(open("single_concept_exercises.pk", "rb"))
````

````python
def clean_code_block_or_print(text):
    if text.startswith("```python\n"):
        if text.endswith("```"):
            text = text[len("```python\n"):-len("```")]
        else:
            text = text[len("```python\n"):].split("```")[0]
    else:
        print(text)
    return text
````

````python
result = []

for sample in data:
    sample = sample.split("PROBLEM:\n")[1]
    problem, solution = sample.split("SOLUTION:\n")
    problem, solution = problem.strip(), solution.strip()
    problem = clean_code_block_or_print(problem)
    solution = clean_code_block_or_print(solution)
    result.append({'instruction': problem, 'input': '', 'output': solution})
    
    
````

````python
for i, r in enumerate(result):
    print(f"{i} problem:")
    print(r['instruction'])
    print("solution:")
    print(r['output'])
    print("#########################")
````

````python
duplicates = [1, 3, 4, 6, 7, 8, 9, 21, 22, 30, 37, 38, 39, 41, 42, 43, 45, 46, 47, 48, 49]
errors = [12, 13, 15, 19, 24, 26, 48, 49]
````

````python
good_examples = [x for i, x in enumerate(result) if i not in duplicates+errors]
````

````python
"""12, 13, 15, 19 are duplicated examples, just fixing one and add to training data"""
fixed_solution = result[12].copy()
fixed_solution['output'] = """
    query_str = f"{column} {operator} @value"
    filtered_df = df.query(query_str)
"""
good_examples.append(fixed_solution)
````

````python
"""24, 26 have typo errors"""
for i in [24, 26]:
    fixed_solution = result[i].copy()
    fixed_solution['output'] = fixed_solution['output'].replace("'col1'", "col1")
    fixed_solution['output'] = fixed_solution['output'].replace("'col2'", "col2")
    good_examples.append(fixed_solution)
````

````python
"""fix indent before <FILL_ME>"""
def reformat_prompt(prompt):
    prompt = "<fim_prefix>\n" + prompt + "\n<fim_middle>"
    return prompt.replace("<FILL_ME>", "<fim_suffix>")

final_solution = []
count = 0
for s in good_examples:
    c_s = s.copy()
    if '\n    <FILL_ME>' in c_s['instruction']:
        count += 1
        c_s['instruction'] = c_s['instruction'].replace('\n    <FILL_ME>', '\n<FILL_ME>')
    c_s['instruction'] = reformat_prompt(c_s['instruction'])
    final_solution.append(c_s)
````

````python
print("% of samples fixed:")
print(f"{count/len(good_examples)*100}%")
````

````python
for i, r in enumerate(final_solution):
    print(f"{i} problem:")
    print(r['instruction'])
    print("solution:")
    print(r['output'])
    print("#########################")
````

````python
import json
output_path = "single_concept_train.json"
with open(output_path, 'w') as json_file:
    json.dump(final_solution, json_file, indent=4)
````
